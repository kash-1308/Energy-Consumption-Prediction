```python
data['HVACUsage'] = data['HVACUsage'].map({'On': 1, 'Off': 0})
data['LightingUsage'] = data['LightingUsage'].map({'On': 1, 'Off': 0})
data['Holiday'] = data['Holiday'].map({'Yes': 1, 'No': 0})
data['DayOfWeek'] = data['DayOfWeek'].astype('category').cat.codes
```

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
data['Year'] = data['Timestamp'].dt.year
data['Month'] = data['Timestamp'].dt.month
data['Day'] = data['Timestamp'].dt.day
data['Hour'] = data['Timestamp'].dt.hour
data['Minute'] = data['Timestamp'].dt.minute
data['Second'] = data['Timestamp'].dt.second
```

```python
data = data.drop(columns=['Timestamp'])
```

```python
col = ['Temperature', 'Humidity', 'SquareFootage', 'Occupancy', 'EnergyConsumption']
plt.figure(figsize=(15, 10))
for i, column in enumerate(col, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=data[column], color='skyblue')
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

```python
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
numerical_columns = ['Temperature', 'Humidity', 'SquareFootage', 'Occupancy', 'RenewableEnergy', 'EnergyConsumption']
data = remove_outliers(data, numerical_columns)
```

```python
col = ['Temperature', 'Humidity', 'SquareFootage', 'Occupancy', 'EnergyConsumption']
plt.figure(figsize=(15, 10))
for i, column in enumerate(col, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=data[column], color='skyblue')
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
X = data.drop(columns=['EnergyConsumption'])
y = data['EnergyConsumption']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=53)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=87)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Display some predictions vs actual values
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(comparison.head(10))
```

```python
from sklearn.metrics import accuracy_score
import numpy as np

# Define a tolerance level, e.g., 10% of the actual value
tolerance = 0.10

# Calculate the percentage of predictions that are within the tolerance level of the actual values
def regression_accuracy(y_true, y_pred, tolerance):
    # Calculate the absolute percentage error
    errors = np.abs((y_true - y_pred) / y_true)
    # Determine if the error is within the tolerance level
    accurate_predictions = errors <= tolerance
    # Calculate the accuracy
    accuracy = np.mean(accurate_predictions)
    return accuracy

# Calculate accuracy with 10% tolerance
accuracy = regression_accuracy(y_test, y_pred, tolerance)

print(f"Custom Regression Accuracy (within {tolerance*100}% tolerance): {accuracy*100:.2f}%")
```
